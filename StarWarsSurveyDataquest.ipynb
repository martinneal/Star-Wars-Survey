# Read in Data and Basic Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

```python
print(star_wars.shape)
star_wars.columns
```

# Cleaning up Respondent ID

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
print(star_wars.shape)
star_wars.head(10)
```

#  Converting Columns to Booleans

```python
yes_no = {'Yes': True , 'No': False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts)
```

```python

    
star_wars = star_wars.rename(columns = {'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
                                       'Unnamed: 4': 'seen_2', 
                                        'Unnamed: 5': 'seen_3', 
                                        'Unnamed: 6': 'seen_4', 
                                        'Unnamed: 7': 'seen_5', 
                                        'Unnamed: 8': 'seen_6'})

for each in star_wars.columns[3:9]:
    print(star_wars[each].value_counts())
```

```python
movies = {'Star Wars: Episode I  The Phantom Menace': True,
         'Star Wars: Episode II  Attack of the Clones': True,
         'Star Wars: Episode III  Revenge of the Sith': True,
         'Star Wars: Episode IV  A New Hope': True,
         'Star Wars: Episode V The Empire Strikes Back': True,
         'Star Wars: Episode VI Return of the Jedi': True,
         np.NaN: False}

for each in star_wars.columns[3:9]:
    star_wars[each] = star_wars[each].map(movies)
    
star_wars.head(5)
```

# Cleaning up Rankings

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

star_wars = star_wars.rename(columns={'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
                                      'Unnamed: 10': 'ranking_2', 
                                      'Unnamed: 11': 'ranking_3', 
                                      'Unnamed: 12': 'ranking_4', 
                                      'Unnamed: 13': 'ranking_5',
                                       'Unnamed: 14': 'ranking_6'})

star_wars[star_wars.columns[9:15]].head(3)
```

# Determining the Highest Ranked Movie

```python
%matplotlib inline

means = star_wars.mean()
```

```python
rankings = ['ranking_1', 'ranking_2', 'ranking_3', 'ranking_4', 'ranking_5', 'ranking_6']
means[rankings].plot.bar()
```

Empire is the highest ranked Star Wars movie in the survey, with Return and New Hope coming in 2nd and 3rd respectably. All three prequel movies rank also 0.5 points higher that the three orginial movies. 

Would like to dig deeper into the age of the participates, & impact only seeing a few movies would impact the rankings. I hypothesize that the older participates are more likely to rank the original movies higher than the new movies. While there is also a chance that some participates may have only viewed the prequel movies. 

# Number of People Who Has Seen Each Movie

```python
sums = star_wars.sum()
sums
```

```python
sums.iloc[3:9].plot.bar()
```

The two most seen movies correspond to the two movies with the highest ranking. Also the lowest ranked movies are the two least seen movies. This indicates a possible correlation between amount seen with ranking. 

```python
plt.scatter(sums.iloc[3:9], means[rankings])
plt.xlabel('Total Views')
plt.ylabel('Ranking (1 to 5)')
```

Scatter plot above only further supports correlation between number of people who have seen the movie and the higher ranking. Unsure about which is the cause and which is the effect.

# Breaking Down Data by Gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

views = ['seen_1', 'seen_2', 'seen_3', 'seen_4', 'seen_5', 'seen_6']
male_sums= males.sum()
male_means= males.mean()
female_sums = females.sum()
female_means = females.mean()

fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.bar(range(6), male_sums[views].values)
ax2.bar(range(6), female_sums[views].values)
ax3.bar(range(6), male_means[rankings].values)
ax4.bar(range(6), female_means[rankings].values)
```

Females have seen fewer Star Wars movies than males, and of the females that have seen Star Wars the proportion which have seen number 2, 3, and 4. Females also rated 1 and 2 higher than their male counterparts, but 4 was rated lower.

# Rankings based on Age

```python
ranking_by_age = star_wars.pivot_table(index='Age', values=rankings, aggfunc=np.mean)
ranking_by_age
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax2 = fig.add_subplot(3,2,2)
ax3 = fig.add_subplot(3,2,3)
ax4 = fig.add_subplot(3,2,4)
ax5 = fig.add_subplot(3,2,5)
ax6 = fig.add_subplot(3,2,6)
ranking_by_age['ranking_1'].plot.bar(ax=ax1, legend=False, ylim=(0,5))
ranking_by_age['ranking_2'].plot.bar(ax=ax2, legend=False, ylim=(0,5))
ranking_by_age['ranking_3'].plot.bar(ax=ax3, legend=False, ylim=(0,5))
ranking_by_age['ranking_4'].plot.bar(ax=ax4, legend=False, ylim=(0,5))
ranking_by_age['ranking_5'].plot.bar(ax=ax5, legend=False, ylim=(0,5))
ranking_by_age['ranking_6'].plot.bar(ax=ax6, legend=False, ylim=(0,5))
```
